# Проект: Предсказание стоимости жилья

**Описание проекта:** Предсказание медианной стоимости жилья в Калифорнии с использованием линейной регрессии.

**Цель проекта:** Разработать и обучить модели машинного обучения для предсказания медианной стоимости дома в жилых массивах Калифорнии на основе различных комбинаций признаков, используя линейную регрессию.

**Описание данных:**

`california_housing_w_features.csv` — данные о жилье в Калифорнии в 1990 году и добавленые новые признаки, сформированные путем преобразования первоначальных данных.

*Данный датасет используется в виду инструкций из 4 темы курса, а так же в чате с преподавателем было сказано использовать его, вопреки инструкциям к работе.*

**Описание признаков:**

- `housing_median_age` — медианный возраст дома в пределах квартала *(меньшее число - это более новое здание)*;
- `total_rooms` — общее количество комнат в домах жилого массива;
- `total_bedrooms` — общее количество спален в домах жилого массива;
- `population` — количество человек, которые проживают в жилом массиве;
- `households` — количество домовладений в жилом массиве;
- `median_income` — медианный доход жителей жилого массива *(единица измерения 10.000 US Dollars)*;
- `median_house_value` — медианная стоимость дома в жилом массиве *(**Целевой признак** / единица измерения US Dollars)*;
- `ocean_proximity` — близость к океану;
- `rooms_per_household` — отношение количества комнат `total_rooms` к количеству домовладений `households`;
- `population_in_household` — отношение количества жителей `population` к количеству домовладений `households`;
- `bedroom_index` — отношение количества спален `total_bedrooms` к общему количеству комнат `total_rooms`.

**Ход исследования:**

- ***Подготовка данных***: загрузка и изучение общей информации из представленных датасетов.
- ***Предобработка данных***: обработка пропущенных значений, корректировка типов данных, дубликатов и других аномалий.
- ***Исследовательский анализ данных***: изучение признаков имеющихся в датасетах, их распределение, поиск выбросов/аномалий в данных.

- ***Подготовка данных:*** кодирование, масштабирование и векторизация разных наборов входных признаков для обучения моделей.

```python
# Функция для анализа количественных значений

def analyzis_quantity(df, name, x_label, y_label='Частота', discrete=False, num=15):
    '''
    Данная функция выводит "коробочный" график и гистограмму 
    по указанному столбцу датафрейма и его статистические метрики.
    Аргументы функции:
    df - данные
    name - название графика
    x_label - подпись для оси Х
    y_label - подпись для оси Y (по умолчанию "Частота")
    discrete - булевое значение, дискретные значение или нет.
    num - количество "корзин" на гистограмме (по умолчанию 15)
    '''
    # Создание составного графика
    fig, (ax_box, ax_hist) = plt.subplots(2, sharex=True, 
                                          figsize=(9, 6), 
                                          gridspec_kw={'height_ratios': (.15, .85)})
    sns.boxplot(x=df, orient='h', ax=ax_box)
    sns.histplot(x=df, bins=num, discrete=discrete, ax=ax_hist)

    # Настройка заголовка и подписей
    ax_box.set(xlabel='')
    fig.suptitle(name, fontsize=14)
    plt.xlabel(x_label)
    plt.ylabel(y_label)
    
    if discrete:
        plt.xticks(np.arange(df.min(), df.max() + 1, 1))
    else:
        pass

    # Отображаем график
    plt.tight_layout()
    plt.show()

    # Вывод стат. метрик столбца
    print(df.describe())
    print()
        
    # Проверяем нормальность распределения
    result = anderson(df)
    print('Тест на нормальность распределения (порог=0.05):')
    print('------------------------------------------------')
    print("Статистика:", result.statistic)
    print("Критические значения:", result.critical_values)
    
    if result.statistic < result.critical_values[2]:
        print('Распределение нормальное.')
    else:
        print('Распределение не является нормальным.')
```

```python
# Функция для поиска выбросов
def outliers(df, col_name):
    '''
    Данная функция предначзначена для 
    поиска выбросов в больших данных.
    На вход функция принимает:
    
    df - переменная DataFrame;
    col_name - имя рассматриваемого признака.
    
    На выходе мы получаем DataFrame c выбросам
    и нижнюю и верхнюю границу, по которым они 
    были выявлены.
    '''
    
    # Вычисляем квартили и IQR
    quantiles = df.approxQuantile(col_name, [0.0, 0.25, 0.5, 0.75, 1.0], 0.01)
    MIN, Q1, Median, Q3, MAX  = quantiles[0], \
                                quantiles[1], \
                                quantiles[2], \
                                quantiles[3], \
                                quantiles[4]
    IQR = Q3 - Q1
    
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    
    print(f"Столбец: {col_name}")
    print('=' * 30)
    print(f"MIN: {MIN:.2f}\nQ1: {Q1:.2f}\nMedian: {Median:.2f}\nQ3: {Q3:.2f}\nMAX: {MAX:.2f}\nIQR: {IQR:.2f}")
    print(f"Границы выбросов: [{lower_bound:.2f}, {upper_bound:.2f}]")
    
    # Находим выбросы
    outliers = df.filter(
        (F.col(col_name) < lower_bound) | (F.col(col_name) > upper_bound))
    
    outlier_count = outliers.count()
    total_count = df.count()
    
    print(f"Количество выбросов: {outlier_count} / {(outlier_count / total_count):.2%} от всех данных.")
    
    return outliers, (lower_bound, upper_bound)
```

```python
# Функция для сравнения распределения случайных выборок
def selection_test(data, col_name, n_selection=4, fraction=0.1, seed=42):
    '''
    Данная функция предназначена для сравнения
    случайных выборок, чтобы оценить, на сколько
    можно им доверять и строить на основе их
    гипотезы.
    
    На ввод функция принимает:
    data - изначальный DataFrame
    col_name - название признака
    n_selection - количество сравниваемых выборок
    fraction - размер выборки
    seed - первоначальный random_state
    '''
    
    fig = plt.subplots(figsize=(10, 5))

    for i in range(n_selection):
        selection = data.select(col_name)\
                        .sample(fraction=0.1, withReplacement=False, seed=seed+69**i )\
                        .toPandas()
        sns.kdeplot(data=selection, label=f'Выборка {i+1}')

    plt.legend()
    plt.title('Сравнение распределений по разным выборкам', fontsize=14)
    plt.ylabel('Плотность')
    plt.xlabel(col_name)
    plt.show()
```

```python
# Функция для анализа количественных признаков
# для работы с Spark
def analyzis_quantity_spark(data, col_name, 
                            n_selection=4, 
                            fraction=0.1, 
                            name=None, 
                            x_label=None, 
                            y_label=None, 
                            num=15,
                            discrete=False,
                            random_state=RANDOM_STATE):
    '''
    Данная функция объединяет инструменты анализа
    количественных признаков:
    outliers(), selection_test(), analyzis_quantity().
    
    На вход функция принимает:
    
    data - изначальный DataFrame
    col_name - название признака
    n_selection - количество сравниваемых выборок
    fraction - размер выборки
    name - Название для итогового графика
    x_label - название оси X
    y_label - название оси Y
    num - количество корзин на гистограмме
    discrete - дискретный признак или непрерывный
    random_state - фиксирование рандомайзера.
    
    На выходе мы получаем DataFrame c выбросам
    и нижнюю и верхнюю границу, по которым они 
    были выявлены.
    '''
    
    # Поиск выбросов
    outlier, bound = outliers(data, col_name)
    
    # Сравнение распределений
    selection_test(data, col_name, 
                   n_selection=n_selection, 
                   fraction=fraction, 
                   seed=random_state)
    
    # Выборка для анализа
    selection = data.select(col_name)\
                    .sample(fraction=fraction, 
                            withReplacement=False, 
                            seed=random_state)\
                    .toPandas()
    # Строим график для оценки распределения
    analyzis_quantity(selection[col_name], 
                      name=name, 
                      x_label=x_label, 
                      y_label=y_label, 
                      discrete=discrete,
                      num=num)
    return outlier, bound
```

```python
# Построение таблицы с расчитанным VIF
def vif_factor(data, columns):
    '''
    Данная функция расчитываем VIF-фактор
    для количественных признаков датафрейма.
    На ввод функция принимает:
    data - датафрейм
    columns - список количественных признаков.
    '''
    vif_data = pd.DataFrame()
    vif_data['input attribute'] = columns
    vif_data['vif'] = ([variance_inflation_factor(data[columns].values, i) 
                        for i in range(data[columns].shape[1])])
    return vif_data
```

```python
# Функция для обучения моделей
def linear_model_spark(data, feature, target, predict):
    '''
    Данная функция предназначена для обучения
    модели линейной регрессии с помощью
    кросс-валидации, подбора гиперпараметров
    и ее оценки на метрике RMSE.
    
    На ввод функция принимает:
    data - тренировочная выборка
    feature - признак с обучающими данными
    target - признак с целевым признаком
    predict - колонка для предсказаний
    
    На выводе мы получаем лучшую модель.
    '''
    # Инициализируем модель
    lr = LinearRegression(featuresCol=feature, 
                          labelCol=target, 
                          predictionCol=predict)
    
    # Сетка гиперпараметров
    param_grid = ParamGridBuilder().addGrid(lr.regParam, [0.01, 0.1, 1])\
                                   .addGrid(lr.elasticNetParam, [0, 0.25, 0.5, 0.75, 1])\
                                   .build()
    
    # Оценка модели
    evaluator = RegressionEvaluator(labelCol=target, 
                                    predictionCol=predict, 
                                    metricName='rmse')
    
    # Кросс-валидация
    cv = CrossValidator(estimator=lr, 
                        estimatorParamMaps=param_grid, 
                        evaluator=evaluator, 
                        numFolds=3, 
                        seed=RANDOM_STATE)
    
    # Обучение модели
    cv_model = cv.fit(data)
    
    # Лучшая модель
    model = cv_model.bestModel
    
    # Вывод результатов
    print('Лучшая метрика RMSE:', min(cv_model.avgMetrics))
    print()
    print("Лучшие параметры:")
    print("regParam =", model.getRegParam())
    print("elasticNetParam =", model.getElasticNetParam())
    
    return model
```

```python
# Функция для оценки модели линейной регресси
def lr_evaluator(data, model, model_name, description, target, predict):
    '''
    Данная функция предназнаена для оценки модели
    линейной регрессии на тестовой выборке.
    На вход модель принимает:
    data - тестовая выборка
    model - оцениваемая модель
    model_name - название модели
    description - описание модели
    target - название целевого признака
    predict - название для колонки с предсказанием
    
    На выходе получаем словарь с результатами.
    '''
    # Предсказание модели
    predictions = model.transform(data)
    
    # Инициализация оценки модели
    evaluator = RegressionEvaluator(labelCol=target, 
                                    predictionCol=predict)
    
    # Оценка модели
    metrics = {
        'model': model_name, 
        'desсription': description,
        'RMSE': evaluator.setMetricName("rmse").evaluate(predictions),
        'MAE': evaluator.setMetricName("mae").evaluate(predictions),
        'R2': evaluator.setMetricName("r2").evaluate(predictions)
    }
    
    return metrics
```

- Полностью на всех данных (по заданию учебного проекта);
- На всех данных кроме категориальных (по заданию учебного проекта);
- На очищенном датасете от излишних и преобразованных признаков для избежания мультиколлинеарности;
- Данные из предыдущего пункта, но без жилых массивов с медианной стоимостью ***500000 US Dollars*** или у которых медианный возраст ***52 года***, так как есть предположение, что это может помешать модели обобщать данные.

### Добавление преобразованных признаков:

```python
df = df.withColumn('population^2', F.col('population')**2)\
       .withColumn('households^2', F.col('households')**2)
```

### Подготовка списков признаков:

```python
# Категориальные признаки
cat_col = ['ocean_proximity']

# Количественные признаки для 1 и 2 набора данных
num_col_1_2 = ['housing_median_age', 'total_rooms', 'total_bedrooms',
               'population', 'households', 'median_income',
               'rooms_per_household', 'population_in_household',
               'bedroom_index']

# Количественные признаки для 3 и 4 набора данных
num_col_3_4 = ['housing_median_age', 'population^2', 'households^2', 
               'median_income', 'bedroom_index', 'rooms_per_household']

# Целевой признак
target = 'median_house_value'
```

### Разбивка на обучающий и тестовый наборы данных:

```python
# Деление на обущающий и тестовый наборы данных
train_data, test_data = df.randomSplit([0.8, 0.2], seed=RANDOM_STATE)
# Подсчет количества строк в подвыборках
print(train_data.count(), test_data.count())
```

#### Фильтрация обучающего набора данных от аномалий и выбросов в данных:

```python
train_data = train_data.filter(F.col('total_rooms') <= 14500)\
                       .filter(F.col('total_bedrooms') <= 3000)\
                       .filter(F.col('population') <= 9000)\
                       .filter(F.col('households') <= 3000)\
                       .filter(F.col('rooms_per_household') <= 25)\
                       .filter(F.col('population_in_household') <= 8)\
                       .filter(F.col('bedroom_index') <= 0.6)
```

```python
# Подсчет количества строк
print(train_data.count())
```

После фильтрации в данных стало меньше строк на **~1.5%**, что является допустимым значением.

#### Фильтрация для 4 обучающего набора данных:

```python
train_data_4 = train_data.filter((F.col('median_house_value') < 500000) & (F.col('housing_median_age') < 52))
```

```python
# Подсчет количества строк
print(train_data_4.count())
```

В данном наборе данных стало меньше на **1630** строк.
